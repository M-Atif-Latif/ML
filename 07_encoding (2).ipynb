## Feature Encoding
Feature encoding is the process of transforming categorical features into numeric features. This is necessary because machine learning algorithms can only handle numeric features. There are many different ways to encode categorical features, and each method has its own advantages and disadvantages. In this notebook, we will explore some of the most popular methods for encoding categorical features, such as:

- Label encoding
- Ordinal encoding
- One-hot encoding
- Binary encoding
- Manual Encoding

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data load
df = sns.load_dataset('tips')
df.head()
```

```python
df['time'].value_counts()
```

## 1. Label Encoding

```python
# lets encode the time column with label encoding of sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
le = LabelEncoder()
df['time'] = le.fit_transform(df['time'])
# lets encode
# the time column with one hot encoding of sklearn
```

```python
df.head()
```

```python
df['day'].value_counts()
```

## 2. Ordinal Encoding

```python
# ordinal encoding the day column using specific order
day_order = ['Thur', 'Fri', 'Sat', 'Sun']
oe = OrdinalEncoder(categories=[day_order])
df['day'] = oe.fit_transform(df[['day']])
df.head()
```

```python
df['day'].value_counts()
```

## 3. One hot encoding

```python
df.head()
```

```python
# one hot encoding on day column
ohe = OneHotEncoder()
encoded_sex = ohe.fit_transform(df[['sex']]).toarray()
```

## 5. Manual Encoding

```python
# manually encoding the time column using pandas
df['time'] = df['time'].map({0: 'Lunch', 1: 'Dinner'})
# manually encoding the day column using pandas
df['day'] = df['day'].map({0: 'Thur', 1: 'Fri', 2: 'Sat', 3: 'Sun'})
# show the encoded colunms
df.head()
```
