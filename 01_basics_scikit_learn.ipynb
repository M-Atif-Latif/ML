# Basic Scikit learn

```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
# load the dataset
df= sns.load_dataset('tips')
# print the first 5 rows of the dataset
df.head()
```

```python
# create a regression line plot using seaborn
sns.lmplot(x='total_bill', y='tip', data=df, ci=None, line_kws={'color': 'red'}, scatter_kws={'color': 'blue'})
plt.title('Regression Line Plot using Seaborn')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.lmplot(x='size', y='tip', data=df)
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
# select the feature colunms from df
X = df[['total_bill']]
# select the target column from df
y = df['tip']
```

```python
# let's create a linear regression model
model = LinearRegression()
# fit the model to the data
model.fit(X, y)
```

```python
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# let's predict the target values using the model
model.predict([[198.524]])
```
